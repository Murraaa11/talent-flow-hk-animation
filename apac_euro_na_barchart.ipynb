```python
gba_in = gba[gba['target_ISO']=='GBA']
gba_out = gba[gba['source_ISO']=='GBA']
```

```python
gba_in_list = []
gba_in_list.append(gba_in[gba_in['source_ISO'].isin(apac_alpha3)].value.sum())
gba_in_list.append(gba_in[gba_in['source_ISO'].isin(euro_alpha3)].value.sum())
gba_in_list.append(gba_in[gba_in['source_ISO'].isin(north_alpha3)].value.sum())
gba_in_list
```

```python
gba_out_list = []
gba_out_list.append(gba_out[gba_out['target_ISO'].isin(apac_alpha3)].value.sum())
gba_out_list.append(gba_out[gba_out['target_ISO'].isin(euro_alpha3)].value.sum())
gba_out_list.append(gba_out[gba_out['target_ISO'].isin(north_alpha3)].value.sum())
gba_out_list
```

```python
sba_in = sba[sba['target_ISO']=='SBA']
sba_out = sba[sba['source_ISO']=='SBA']
```

```python
sba_in_list = []
sba_in_list.append(sba_in[sba_in['source_ISO'].isin(apac_alpha3)].value.sum())
sba_in_list.append(sba_in[sba_in['source_ISO'].isin(euro_alpha3)].value.sum())
sba_in_list.append(sba_in[sba_in['source_ISO'].isin(north_alpha3)].value.sum())
sba_in_list
```

```python
sba_out_list = []
sba_out_list.append(sba_out[sba_out['target_ISO'].isin(apac_alpha3)].value.sum())
sba_out_list.append(sba_out[sba_out['target_ISO'].isin(euro_alpha3)].value.sum())
sba_out_list.append(sba_out[sba_out['target_ISO'].isin(north_alpha3)].value.sum())
sba_out_list
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 數據示例
regions = ['APAC', 'Europe', 'America']

# 繪製柱狀圖
bar_width = 0.2
x = np.arange(len(regions))

fig, ax = plt.subplots(figsize=(6, 4))

# 繪製兩組柱狀
bars1 = ax.bar(x - bar_width/2, gba_in_list, bar_width, label='Great Bay Area', color='#00a8e8')
bars2 = ax.bar(x + bar_width/2, sba_in_list, bar_width, label='San Francisco Bay Area', color='#ed3b9c')

# 添加標籤和標題
# ax.set_xlabel('地區', fontsize=14)
ax.set_ylabel('No. of Researchers Inflow', fontsize=14)
ax.set_title('Mobility Inflow', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(regions, fontsize=12)
ax.legend(fontsize=10)

# 添加數據標籤
def add_data_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', fontsize=10)

add_data_labels(bars1)
add_data_labels(bars2)

# 隐藏纵坐标和边框
ax.spines['left'].set_visible(False)  # 隐藏左边框
ax.spines['right'].set_visible(False)  # 隐藏右边框
ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.get_yaxis().set_visible(False)  # 隐藏纵坐标

# 顯示網格
ax.yaxis.grid(False)

# 顯示圖形
plt.tight_layout()
plt.savefig('./inflow.png', bbox_inches='tight', dpi=300)
plt.show()
```
